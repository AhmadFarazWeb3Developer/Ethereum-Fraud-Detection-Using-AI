## II. Dataset Loading

```python
import pandas as pd

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Cleaned_Ethereum_Fraud_Detection.csv')
df.head()
```

## III. Dataset Overview

## IV. Descriptive Statistics

```python
df.describe()
```

## V. Target Variable Analysis (FLAG)

```python
df['FLAG'].value_counts()
```

```python
sns.countplot(x='FLAG', data=df)
plt.title('Class Distribution (FLAG)')
plt.show()
```

## VI. Distribution Analysis

```python
df.hist(figsize=(15,12))
plt.show()
```

## VII. Outlier Detection using Boxplots

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df.select_dtypes(include=np.number))
plt.xticks(rotation=90)
plt.show()
```

## VIII. Correlation Analysis

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(14,10))
sns.heatmap(
    numeric_df.corr(),
    cmap='coolwarm',
    linewidths=0.5
)
plt.title("Correlation Heatmap (Numeric Features Only)")
plt.show()
```

## IX. Relationship Analysis

```python
sns.scatterplot(
    x='total transactions (including tnx to create contract',
    y='total ether received',
    hue='FLAG',
    data=df,
    alpha=0.6
)

plt.title("Total Transactions vs Total Ether Received")
plt.show()
```
